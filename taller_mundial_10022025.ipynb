# üìÇ PASO 1: Cargar los datos

````python
import pandas as pd
import numpy as np
import sys

# Configuraci√≥n para ver m√°s datos
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("‚úÖ Librer√≠as cargadas")
print(f"   Python: {sys.version.split()[0]}")
print(f"   Pandas: {pd.__version__}")
````

````python
# Encontrar y cargar los archivos CSV
import os

ruta_lesiones = None
ruta_jugadores = None

for root, dirs, files in os.walk('football-datasets'):
    for f in files:
        if f == 'player_injuries.csv':
            ruta_lesiones = os.path.join(root, f)
        if f == 'player_profiles.csv':
            ruta_jugadores = os.path.join(root, f)

print(f"üìÇ Archivos encontrados:")
print(f"   Lesiones: {ruta_lesiones}")
print(f"   Jugadores: {ruta_jugadores}")
````

````python
# Cargar datos de lesiones
df_lesiones = pd.read_csv(ruta_lesiones)

print(f"üî• DATOS REALES DE LESIONES: {len(df_lesiones):,} registros")
print(f"   Columnas: {list(df_lesiones.columns)}")
````

````python
# Cargar datos de jugadores
df_jugadores = pd.read_csv(ruta_jugadores)

print(f"üî• DATOS REALES DE JUGADORES: {len(df_jugadores):,} registros")
print(f"   Columnas: {list(df_jugadores.columns)}")
````

````python
# Unir los datos: lesiones + informaci√≥n del jugador
# Primero veamos qu√© columna tienen en com√∫n
print("Columnas en lesiones:", df_lesiones.columns.tolist())
print("Columnas en jugadores:", df_jugadores.columns.tolist())
````

````python
# Unir por player_id
columnas_jugador = ['player_id', 'player_name', 'citizenship', 'position', 'foot']
# Filtrar solo las columnas que existen
cols_disponibles = [c for c in columnas_jugador if c in df_jugadores.columns]

df = df_lesiones.merge(
    df_jugadores[cols_disponibles],
    on='player_id',
    how='left'
)

print(f"\nüèÜ DATOS UNIDOS: {len(df):,} registros de lesiones REALES")
print(f"\nüá®üá¥ ¬°Faltan 4 meses para el Mundial 2026!")
print(f"\nüìä Fuente: Transfermarkt (datos reales de jugadores profesionales)")
````

````python
from docx import Document
from datetime import datetime
import pandas as pd

doc = Document()

doc.add_heading('INFORME M√âDICO - SELECCI√ìN COLOMBIA 2026', level=0)
doc.add_paragraph(f'Fecha de generaci√≥n: {datetime.today().strftime("%d/%m/%Y")}')

def agregar_codigo(texto):
    p = doc.add_paragraph()
    run = p.add_run(texto)
    run.font.name = 'Courier New'

def dataframe_a_tabla(df_tabla):
    tabla = doc.add_table(rows=df_tabla.shape[0]+1, cols=df_tabla.shape[1])
    for j, col in enumerate(df_tabla.columns):
        tabla.rows[0].cells[j].text = str(col)
    for i in range(df_tabla.shape[0]):
        for j in range(df_tabla.shape[1]):
            tabla.rows[i+1].cells[j].text = str(df_tabla.iloc[i,j])

def serie_a_tabla(serie):
    df_tabla = serie.reset_index()
    df_tabla.columns = ['Valor','Cantidad']
    dataframe_a_tabla(df_tabla)

doc.add_heading('Fase 1: Exploraci√≥n Libre de Datos', level=1)

doc.add_heading('df.head()', level=2)
agregar_codigo("df.head()")
dataframe_a_tabla(df.head())

doc.add_heading('df.info()', level=2)
agregar_codigo("df.info()")
info_df = pd.DataFrame({
    "Columna": df.columns,
    "No Nulos": df.notnull().sum().values,
    "Tipo": df.dtypes.values
})
dataframe_a_tabla(info_df)

doc.add_heading("df['days_missed'].describe()", level=2)
agregar_codigo("df['days_missed'].describe()")
serie_a_tabla(df['days_missed'].describe())

doc.add_heading("df['foot'].value_counts()", level=2)
agregar_codigo("df['foot'].value_counts()")
serie_a_tabla(df['foot'].value_counts())

peor_posicion = posicion_riesgo.index[0]
lesion_mas_frecuente_mundial = top_10_lesiones.index[0]
lesion_mas_frecuente_colombia = top_lesiones_col.index[0]
pais_mas_lesionado_grupo = resumen_final.index[0]
promedio_partidos_grupo = resumen_final['promedio_partidos'].mean()

doc.add_heading('Resumen Ejecutivo', level=1)

doc.add_paragraph(
    "Se realiz√≥ un an√°lisis epidemiol√≥gico sobre lesiones en futbolistas profesionales "
    "utilizando datos hist√≥ricos reales provenientes de Transfermarkt."
)

doc.add_heading('Hallazgos Globales', level=2)

agregar_codigo("df.shape[0]")
doc.add_paragraph(f"Total de lesiones registradas: {total_lesiones:,}")

agregar_codigo("df['injury_reason'].value_counts().head(10)")
serie_a_tabla(top_10_lesiones)

doc.add_paragraph(
    f"La lesi√≥n m√°s frecuente a nivel mundial es '{lesion_mas_frecuente_mundial}'."
)

agregar_codigo("df['days_missed'].mean(), df['days_missed'].max()")
doc.add_paragraph(
    f"Promedio de incapacidad: {promedio_dias:.1f} d√≠as | M√°ximo hist√≥rico: {max_dias} d√≠as."
)

doc.add_heading('An√°lisis Espec√≠fico Selecci√≥n Colombia', level=2)

agregar_codigo("df[df['citizenship'] == 'Colombia']")
doc.add_paragraph(f"Lesiones de jugadores colombianos: {len(df_colombia)}")

agregar_codigo("df_colombia['injury_reason'].value_counts().head(5)")
serie_a_tabla(top_lesiones_col)

doc.add_paragraph(
    f"La lesi√≥n predominante en Colombia es '{lesion_mas_frecuente_colombia}'."
)

agregar_codigo("df['position'].value_counts()")
serie_a_tabla(posicion_riesgo.head(10))

doc.add_paragraph(
    f"La posici√≥n con mayor incidencia de lesi√≥n es '{peor_posicion}'."
)

doc.add_heading('Lesiones M√°s Graves (Mayor tiempo de recuperaci√≥n)', level=2)

agregar_codigo("df.groupby('injury_reason')['days_missed'].mean().sort_values(ascending=False).head(10)")
serie_a_tabla(lesiones_graves)

doc.add_paragraph(
    "Estas lesiones representan el mayor riesgo competitivo por tiempo prolongado fuera de competencia."
)

doc.add_heading('Jugadores Colombianos con Mayor Historial de Lesiones', level=2)

agregar_codigo("df_colombia['player_name'].value_counts().head(10)")
serie_a_tabla(top_jugadores_col)

doc.add_paragraph(
    "Estos jugadores requieren monitoreo de carga f√≠sica y planes preventivos individualizados."
)


doc.add_heading('Conclusi√≥n', level=2)

doc.add_paragraph(
    "El an√°lisis evidencia patrones predecibles por posici√≥n y tipo de esfuerzo, "
    "permitiendo dise√±ar prevenci√≥n basada en datos para el Mundial 2026."
)


doc.save("Informe_Medico_Seleccion_Colombia.docx")


````
