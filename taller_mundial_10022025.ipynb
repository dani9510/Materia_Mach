# ðŸ“‚ PASO 1: Cargar los datos

````python
import pandas as pd
import numpy as np
import sys

# ConfiguraciÃ³n para ver mÃ¡s datos
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("âœ… LibrerÃ­as cargadas")
print(f"   Python: {sys.version.split()[0]}")
print(f"   Pandas: {pd.__version__}")
````

````python
# Encontrar y cargar los archivos CSV
import os

ruta_lesiones = None
ruta_jugadores = None

for root, dirs, files in os.walk('football-datasets'):
    for f in files:
        if f == 'player_injuries.csv':
            ruta_lesiones = os.path.join(root, f)
        if f == 'player_profiles.csv':
            ruta_jugadores = os.path.join(root, f)

print(f"ðŸ“‚ Archivos encontrados:")
print(f"   Lesiones: {ruta_lesiones}")
print(f"   Jugadores: {ruta_jugadores}")
````

````python
# Cargar datos de lesiones
df_lesiones = pd.read_csv(ruta_lesiones)

print(f"ðŸ”¥ DATOS REALES DE LESIONES: {len(df_lesiones):,} registros")
print(f"   Columnas: {list(df_lesiones.columns)}")
````

````python
# Cargar datos de jugadores
df_jugadores = pd.read_csv(ruta_jugadores)

print(f"ðŸ”¥ DATOS REALES DE JUGADORES: {len(df_jugadores):,} registros")
print(f"   Columnas: {list(df_jugadores.columns)}")
````

````python
# Unir los datos: lesiones + informaciÃ³n del jugador
# Primero veamos quÃ© columna tienen en comÃºn
print("Columnas en lesiones:", df_lesiones.columns.tolist())
print("Columnas en jugadores:", df_jugadores.columns.tolist())
````

````python
# Unir por player_id
columnas_jugador = ['player_id', 'player_name', 'citizenship', 'position', 'foot']
# Filtrar solo las columnas que existen
cols_disponibles = [c for c in columnas_jugador if c in df_jugadores.columns]

df = df_lesiones.merge(
    df_jugadores[cols_disponibles],
    on='player_id',
    how='left'
)

print(f"\nðŸ† DATOS UNIDOS: {len(df):,} registros de lesiones REALES")
print(f"\nðŸ‡¨ðŸ‡´ Â¡Faltan 4 meses para el Mundial 2026!")
print(f"\nðŸ“Š Fuente: Transfermarkt (datos reales de jugadores profesionales)")
````

````python
# Top 10 lesiones: % del total, promedios y top5 de citizenship/position/foot
import pandas as pd

def top_n_concat(series, n=5, sep='; '):
    vals = series.dropna().astype(str).value_counts().head(n).index.tolist()
    return sep.join(vals)

# Conteo top 10 lesiones
top10_counts = df['injury_reason'].value_counts().head(10)
top10 = top10_counts.rename_axis('injury_reason').reset_index(name='count')

total = len(df)

# Promedios por injury_reason (puede haber NaN)
aggs = df.groupby('injury_reason').agg(
    avg_days_missed = ('days_missed', 'mean'),
    avg_games_missed = ('games_missed', 'mean')
).reset_index()

# Top5 concatenados por categorÃ­a
citizenship_top5 = df.groupby('injury_reason')['citizenship'].apply(lambda s: top_n_concat(s, 5)).reset_index(name='top5_citizenship')
position_top5 = df.groupby('injury_reason')['position'].apply(lambda s: top_n_concat(s, 5)).reset_index(name='top5_position')
foot_top5 = df.groupby('injury_reason')['foot'].apply(lambda s: top_n_concat(s, 5)).reset_index(name='top5_foot')

# Combinar todo y filtrar solo top10
result = top10.merge(aggs, on='injury_reason', how='left') \
               .merge(citizenship_top5, on='injury_reason', how='left') \
               .merge(position_top5, on='injury_reason', how='left') \
               .merge(foot_top5, on='injury_reason', how='left')

# AÃ±adir porcentaje del total
result['pct_total'] = result['count'] / total * 100

# Formateo: redondear y reordenar columnas
result['pct_total'] = result['pct_total'].round(2)
result['avg_days_missed'] = result['avg_days_missed'].round(2)
result['avg_games_missed'] = result['avg_games_missed'].round(2)

cols = ['injury_reason','count','pct_total','avg_days_missed','avg_games_missed',
        'top5_citizenship','top5_position','top5_foot']
result = result[cols]

print("Top 10 lesiones mundiales:")
print(result.to_string(index=False))
````
